# MQAR Experiment Analysis

This notebook analyzes the results of the Multi-Query Associative Recall (MQAR) experiments.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mqar_analysis import load_results, plot_results

%matplotlib inline
```

```python
# Load Results
OUT_ROOT = "./out/mqar"
SAVE_DIR = "./analysis_results"

if not os.path.exists(OUT_ROOT):
    print(f"Warning: {OUT_ROOT} does not exist. Please run experiments first.")
    # Create dummy data for visualization demonstration if needed
    pass
else:
    df = load_results(OUT_ROOT)
    if not df.empty:
        display(df.head())
        
        # Save summary
        os.makedirs(SAVE_DIR, exist_ok=True)
        df.to_csv(os.path.join(SAVE_DIR, "summary.csv"), index=False)
    else:
        print("No results found in", OUT_ROOT)
```

```python
# Plot Results
if 'df' in locals() and not df.empty:
    plot_results(df, SAVE_DIR)
    
    # Additional custom analysis can be done here
    pivot = df.pivot_table(index='Model', columns=['Seq Len', 'Key Len'], values='Test Acc')
    display(pivot)
```
